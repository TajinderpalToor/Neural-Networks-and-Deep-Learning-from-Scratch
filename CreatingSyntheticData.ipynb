In this program, we are going to look at creating synthetic data. Using numopy to create and plot random data points so that we can perform an algorithm on it.

The data  is created in the shape of rectangle, circle or even a sphere. Really it can be anything we want it to be.

Once we create this syntehtic data and label it. We can perform the basic perceptron algo. We will complete it with circles, the assigment is with spheres

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Example 1: Create 1000 points uniformly distributed in [7, 11] Ã— [5, 8].

# [7,11] is the x cordinates and [5,8] is the y cordinates
#np.random.rand uses a normal/gaussian distribution

x1= 7 + 4*np.random.rand(1,300)
x2= 5 + 3*np.random.rand(1,300)
plt.xlim(0,15)
plt.ylim(0,15)
plt.scatter(x1[0],x2[0], color='green',s=10)
plt.show()
```

```python
# Example 2: Create points in a circle -- uniform distribution with a radius of 3, centered at [7,5]

radius= 3*np.random.rand(1,3000)
theta=2*np.pi*np.random.rand(1,3000)
x1=7+radius*np.cos(theta)
x2=5+radius*np.sin(theta)

plt.xlim(0,11)
plt.ylim(0,11)
plt.scatter(x1[0],x2[0], color='green',s=1)
plt.show()
```

```python
# Example 2: Create points in a circle -- normal distribution with a radius of 3, centered at [15,15]

radius= 3*np.random.normal(0,1,size=(1,3000))
theta=2*np.pi*np.random.normal(0,1,size=(1,3000))
x1=15+radius*np.cos(theta)
x2=15+radius*np.sin(theta)
plt.xlim(0,25)
plt.ylim(0,25)

plt.scatter(x1[0],x2[0], color='green',s=1)
plt.show()
```

```python
# Example 3: Create 1000 points uniformly distributed in [7, 11] Ã— [5, 8]Ã— [5, 8] cube.

# [7,11] is the x cordinates and [5,8] is the y cordinates
#np.random.rand uses a UNIFORM distribution

from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


x1= 7 + 4*np.random.rand(1,30000)
x2= 5 + 3*np.random.rand(1,30000)
x3= 5 + 3*np.random.rand(1,30000)


ax.scatter(x1[0],x2[0],x3[0], color='green',s=1)
plt.show()
```
