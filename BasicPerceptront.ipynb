```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
N=1000 # number of points per cluster
M=20 # number of points used for training
p= [6,13,16,5] # cordinates of the centes of the clusters -- 6,13 is one and 16,5 is the other
```

```python
x=np.zeros((2*N,2)) # creating an array of zeros (2000 becasue there is that many points, 2 coloums cause there is an xy for each point)
t=np.zeros((2*N,1)) # creating an array of 2000 points and one row
```

```python
#Create the synthetic data -- using circles, normally distributed points and the theta of the circle is uniformly distributed

# rememebr array x[0] has the x  and x[1] has the y of our point
#p[0] = x cordinate of the first cluster and p[1] is the y cordinate of the zcluster

radius= np.random.normal(0,1,size=2*N)
theta=2*np.pi*np.random.rand(2*N)

x[0:N,0]=p[0]+radius[0:N]*np.cos(theta[0:N])
x[0:N,1]=p[1] +radius[0:N]*np.sin(theta[0:N])
t[0:N]=0

x[N:2*N,0]=p[2]+radius[N:2*N]*np.cos(theta[N:2*N])
x[N:2*N,1]=p[3] +radius[N:2*N]*np.sin(theta[N:2*N])
t[N:2*N]=1

plt.figure(figsize=(10,10))
plt.scatter(x[:,0], x[:,1])
```

```python
# Paramter Initialization -- w1,w2 and b should be a number between -0.5 and 0.5
w1= 0.5-np.random.rand()
w2=0.5-np.random.rand()
b=0.5-np.random.rand()

ii=np.linspace(1,15,15)
#plt.plot(ii, (-b-w1*ii)/w2)
print(type(w2))
print(type(x))

a=range(0,2*N) # range of numbers from 0 to 1999

sp=random.sample(a,M+1000); # indexes for random selection of data for training and testing.
# random sample has two paramters,the first one is the list of numbers, and the second one takes a random list of that size from the list given in the first paramter
# taking 1020 because, 1000 is for testing and 20 is for training
```

```python
# Training

for i in range(0,M):

  #calculate the prediction
  y=b + w1*x[sp[i],0] + w2*x[sp[i],1]
  if y<0:
    y=0
  else:
    y=1

  # calculate the error so that we can update the paramters
  error= t[sp[i]]-y

  # update the paramters
  w1= w1 + error*x[sp[i],0]
  w2= w2 + error*x[sp[i],1]
  b=b+error


```

```python
# Testing
er=0
for i in range(M,M+1000):
  y = b + w1*x[sp[i],0] + w2*x[sp[i],1]
  if y<0:
    y=0
  else:
    y=1
  e = abs(t[sp[i]] - y)
  er = er + e
er=er/1000


#plot the line
plt.plot(ii, (-b-w1*ii)/w2)

#plot the scatter plot
plt.scatter(x[:,0], x[:,1])

plt.show()

print(er)
```
